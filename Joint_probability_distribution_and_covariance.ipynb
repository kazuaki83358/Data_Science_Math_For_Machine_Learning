```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Make joint distribution (2 variables)
X = np.random.normal(0, 1, 1000)
Y = X * 0.8 + np.random.normal(0, 0.3, 1000)  # correlated

sns.jointplot(x=X, y=Y, kind="hex")
plt.show()
```

```python
import numpy as np

X = np.random.normal(0, 1, 1000)
Y = 2*X + np.random.normal(0, 0.5, 1000)

cov = np.cov(X, Y)[0][1]
print("Covariance:", cov)
import matplotlib.pyplot as plt

plt.scatter(X, Y, alpha=0.5)
plt.title("High Covariance Example")
plt.xlabel("X")
plt.ylabel("Y")
plt.grid()
plt.show()
```
