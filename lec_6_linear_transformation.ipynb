```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Original vectors 
x = np.array([[1,0],[0,1]]).T #basis vector (i-hat, j-hat)
```

```python
# Transformation matrix (example: rotation by 45Â°)
theta = np.radians(45)
A = np.array([[np.cos(theta), -np.sin(theta)],
              [np.sin(theta), np.cos(theta)]])
```

```python
#Apply transformation
x_transformed = A @ x
```

```python
# Plotting
plt.figure(figsize=(6,6))
plt.axhline(0, color='gray', linewidth=1)
plt.axvline(0, color='gray', linewidth=1)
```

```python
# Original basis
plt.quiver(0, 0, x[0,0], x[1,0], angles='xy', scale_units='xy', scale=1, color='b', label='Original x-axis')
plt.quiver(0, 0, x[0,1], x[1,1], angles='xy', scale_units='xy', scale=1, color='g', label='Original y-axis')
```

```python
# Transformed basis
plt.quiver(0, 0, x_transformed[0,0], x_transformed[1,0], angles='xy', scale_units='xy', scale=1, color='r', label='Transformed x-axis')
plt.quiver(0, 0, x_transformed[0,1], x_transformed[1,1], angles='xy', scale_units='xy', scale=1, color='orange', label='Transformed y-axis')
plt.xlim(-2, 2)
plt.ylim(-2, 2)
plt.legend()
plt.title("Linear Transformation (Rotation by 45Â°)")
plt.grid(True)
plt.show()
```
