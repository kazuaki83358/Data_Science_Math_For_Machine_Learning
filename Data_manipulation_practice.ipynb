```python
df['Math_Score'].fillna(df['Math_Score'].mean(), inplace=True)
df['Science_Score'].fillna(df['Science_Score'].mean(), inplace=True)
df['English_Score'].fillna(df['English_Score'].mean(), inplace=True)
df['Attendance'].fillna(df['Attendance'].mean(), inplace=True)
```

```python
df
```

```python
df.dropna(subset=['Name', 'Gender'], inplace=True)
```

```python
df
```

```python
df.rename(columns={
    'Math_Score': 'Math',
    'Science_Score': 'Science',
    'English_Score': 'English'
}, inplace=True)
```

```python
df
```

```python
df['Attendance'] = df['Attendance'].astype(int)
```

```python
df.to_csv('student_performance_cleaned.csv', index=False)
```
