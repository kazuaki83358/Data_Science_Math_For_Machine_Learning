```python
#sym.integrate(fanc,var)
#sym.integrate(func(var,lower_limit,upper_limit)
import sympy as sym
x = sym.symbols('x')
integ = sym.integrate(sym.sin(x),x)
print(integ)
```

```python
#sym.integrate(fanc,var)
#sym.integrate(func(var,lower_limit,upper_limit)
import sympy as sym
x = sym.symbols('x')
integ = sym.integrate(sym.sin(x),(x,0,1))
print(integ)
```

```python
import sympy as sym
x = sym.symbols('x')
integ = sym.integrate(sym.exp(-x),(x,0,sym.oo))
print(integ)
```

```python
#double integral
import sympy as sym
x,y = sym.symbols('x y')
dinteg = sym.integrate(sym.exp(-x**2-y**2),(x,-sym.oo,sym.oo),(y,-sym.oo,sym.oo))
print(dinteg)
```

```python
#limits
#sym.limit(func,var,point)
import sympy as sym
x = sym.symbols('x')
limit1=sym.limit(sym.exp(-x),x,0)
print(limit1)
```

```python
import sympy as sym
x = sym.symbols('x')
limit1=sym.limit(sym.exp(-x)/x,x,sym.oo)
print(limit1)
```

```python
#series expansion
#sym.series(func,x,x0,n)
import sympy as sym
x=sym.Symbol('x')
exp = sym.series(sym.cos(x),x)
print(exp)
```

```python
#series expansion
#sym.series(func,x,x0,n)
import sympy as sym
x=sym.Symbol('x')
exp1 = sym.series(sym.cos(x)**2,x,1,4)
print(exp1)
```
