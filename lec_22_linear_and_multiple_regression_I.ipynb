*Itâ€™s the simplest machine learning algorithm â€” it tries to draw a straight line (or plane) that best fits your data.

You use it when you want to predict a continuous value (like price, salary, temperature, etc.).*

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# Data: Study Hours vs Exam Score
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
y = np.array([2, 4, 5, 4, 5])
```

```python
y
```

```python
# Train Model
model = LinearRegression()
model.fit(X, y)
```

```python

# Predict
y_pred = model.predict(X)
```

```python
# Plot
plt.scatter(X, y, color='blue', label='Actual')
plt.plot(X, y_pred, color='red', label='Predicted Line')
plt.xlabel('Hours Studied')
plt.ylabel('Exam Score')
plt.legend()
plt.title('Simple Linear Regression')
plt.show()
```

##Now we have more than one input feature.
Predicting house price using:
size (sqft)
number of rooms
distance from city

```python
import numpy as np
from sklearn.linear_model import LinearRegression
```

```python
# Data: [size, rooms]
X = np.array([[1000, 2],
              [1500, 3],
              [2000, 3],
              [2500, 4],
              [3000, 4]])
```

```python
# Target: house price (in lakhs)
y = np.array([50, 65, 80, 90, 105])
```

```python
# Train Model
model = LinearRegression()
model.fit(X, y)
```

```python
# Predict price of new house
new_house = np.array([[2200, 3]])
pred = model.predict(new_house)

print(f"Predicted Price: {pred[0]:.2f} Lakhs")
```
