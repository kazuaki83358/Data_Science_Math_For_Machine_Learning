```python
# Linear Discriminant Analysis (LDA) Example
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_wine
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
```

```python
# Step 1: Load the Iris dataset
wine = load_wine()
X = wine.data       # features: sepal & petal length/width
y = wine.target     # labels: 0, 1, 2 (Setosa, Versicolor, Virginica)
```

```python
# Step 2: Apply LDA (reduce to 2D for visualization)
lda = LDA(n_components=2)
X_lda = lda.fit_transform(X, y)
```

```python
# Step 3: Plot results
plt.figure(figsize=(8, 6))
colors = ['red', 'green', 'blue']
labels = wine.target_names

for color, i, label in zip(colors, [0, 1, 2], labels):
    plt.scatter(X_lda[y == i, 0], X_lda[y == i, 1], alpha=0.8, color=color, label=label)

plt.xlabel('LDA Component 1')
plt.ylabel('LDA Component 2')
plt.title('Linear Discriminant Analysis on Iris Dataset')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# PCA for comparison
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
result = pca.fit(X)
z = result.transform(X)
plt.scatter(z[:, 0], z[:, 1], c=y)
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.title('PCA on Iris Dataset')    
plt.grid(True)
plt.show()
```
