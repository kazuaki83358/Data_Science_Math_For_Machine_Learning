```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Creating a random matrix (e.g., image-like data)
A = np.random.rand(6, 6)
```

```python
# Perform SVD
U, S, Vt = np.linalg.svd(A)
```

```python
# Reconstruct matrix using top k singular values
def low_rank_approximation(U, S, Vt, k):
    """Reconstruct matrix using only top k singular values"""
    S_k = np.zeros((U.shape[0], Vt.shape[0]))
    np.fill_diagonal(S_k, S[:k])
    A_k = U @ S_k @ Vt
    return A_k
```

```python
# Compare approximations
fig, axes = plt.subplots(1, 4, figsize=(12, 4))
ranks = [1, 2, 3, len(S)]
for i, k in enumerate(ranks):
    A_k = low_rank_approximation(U, S, Vt, k)
    axes[i].imshow(A_k, cmap='viridis')
    axes[i].set_title(f'Rank-{k} Approximation')
    axes[i].axis('off')

plt.suptitle("Low-Rank Matrix Approximation using SVD", fontsize=14)
plt.show()
```
