```python
#Dice PMF
import matplotlib.pyplot as plt

values = [1, 2, 3, 4, 5, 6]
pmf = [1/6]*6

plt.stem(values, pmf)
plt.title("PMF of a Fair Dice")
plt.xlabel("Outcome")
plt.ylabel("Probability")
plt.grid()
plt.show()
```

```python
#PDF (Probability Density Function)
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

x = np.linspace(-4, 4, 400)
y = norm.pdf(x, 0, 1)

plt.plot(x, y)
plt.title("PDF - Normal Distribution")
plt.xlabel("x")
plt.ylabel("Density")
plt.grid()
plt.show()
```

```python
#CDF (Cumulative Distribution Function)
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

x = np.linspace(-4, 4, 400)
y = norm.cdf(x, 0, 1)

plt.plot(x, y)
plt.title("CDF - Normal Distribution")
plt.xlabel("x")
plt.ylabel("F(x)")
plt.grid()
plt.show()
```
