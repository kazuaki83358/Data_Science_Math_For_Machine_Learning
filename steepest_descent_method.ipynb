```python
import numpy as np
import matplotlib.pyplot as plt

def f(x, y):
    return x**2 + y**2

def grad_f(x, y):
    return np.array([2*x, 2*y])

x = np.array([4.0, 3.0])
alpha = 0.1

xs = []
ys = []

for i in range(30):
    xs.append(x[0])
    ys.append(x[1])
    gradient = grad_f(x[0], x[1])
    x = x - alpha * gradient

plt.plot(xs, ys, marker="o")
plt.title("Steepest Descent Path")
plt.xlabel("x")
plt.ylabel("y")
plt.grid()
plt.show()
```
