```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_excel('flight-price.xlsx')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.head()
```

```python
## feature engineering
df['Date'] = df['Date_of_Journey'].str.split('/').str[0]
df['Month'] = df['Date_of_Journey'].str.split('/').str[1]
df['Year'] = df['Date_of_Journey'].str.split('/').str[2]
```

```python
df.head()
```

```python
df.info()
```

```python
df['Date'] = df['Date'].astype(int)
df['Month'] = df['Month'].astype(int)
df['Year'] = df['Year'].astype(int)
```

```python
df.info()
```

```python
#drop date of journey
df.drop('Date_of_Journey', axis=1, inplace=True)
```

```python
df.head()
```

```python
df['Arrival_Time'] = df['Arrival_Time'].apply(lambda x : x.split(' ')[0])
```

```python
df['Arrival_Hour'] = df['Arrival_Time'].str.split(':').str[0]
df['Arrival_Min'] = df['Arrival_Time'].str.split(':').str[1]
```

```python
df.head(2)
```

```python
df['Arrival_Hour'] = df['Arrival_Hour'].astype(int)
df['Arrival_Min'] = df['Arrival_Min'].astype(int)
```

```python
df.drop('Arrival_Time', axis=1, inplace=True)
```

```python
df.info()
```

```python
df['Dep_Hour'] = df['Dep_Time'].str.split(':').str[0]
df['Dep_Min'] = df['Dep_Time'].str.split(':').str[1]
```

```python
df.head(2)
```

```python
df.info()
```

```python
df['Dep_Hour'] = df['Dep_Hour'].astype(int)
df['Dep_Min'] = df['Dep_Min'].astype(int)
```

```python
df.info()
```

```python
df.drop('Dep_Time', axis=1, inplace=True)
```

```python
df.head(5)
```

```python
df['Duration_Hour'] = df['Duration'].str.split(' ').str[0].str.split('h').str[0]
df['Duration_Min'] = df['Duration'].str.split(' ').str[1].str.split('m').str[0]
```

```python
df['Duration_Min'].isna().sum()
```

```python
df['Duration_Min'] = df['Duration_Min'].fillna(0)
```

```python
df['Duration_Min'].isna().sum()
```

```python
df['Duration_Hour'].isna().sum()
```

```python
df['Duration_Hour'] = df['Duration'].str.extract('(\\d+)h', expand=False).fillna('0')
df['Duration_Min'] = df['Duration'].str.extract('(\\d+)m', expand=False).fillna('0')

# Convert both extracted columns to integer type
df['Duration_Hour'] = df['Duration_Hour'].astype(int)
df['Duration_Min'] = df['Duration_Min'].astype(int)
```

```python
df.info()
```

```python
df.head()
```

```python
df.drop('Duration',axis=1,inplace=True)
```

```python
df.head(2)
```

```python
df.info()
```

```python
df['Total_Stops'] = df['Total_Stops'].map({'non-stop':0,'1 stop':1,'2 stops':2,'3 stops':3,'4 stops':4,np.nan:1})
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
encoder = OneHotEncoder()
```

```python
encoder.fit_transform(df[['Airline','Source','Destination','Additional_Info']]).toarray()
```

```python
pd.DataFrame(encoder.fit_transform(df[['Airline','Source','Destination','Additional_Info']]).toarray(),columns=encoder.get_feature_names_out())
```

```python
ohe_df = pd.DataFrame(encoder.fit_transform(df[['Airline','Source','Destination','Additional_Info']]).toarray(),columns=encoder.get_feature_names_out())
df = pd.concat([df, ohe_df], axis=1)
```

```python
df.drop(['Airline','Source','Destination','Additional_Info'],axis=1,inplace=True)
```

```python
df.to_csv('flight_price_encoded.csv', index=False)
print('DataFrame saved to flight_price_encoded.csv')
```
