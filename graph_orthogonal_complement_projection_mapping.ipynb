```python
import numpy as np
import matplotlib.pyplot as plt

# Vector v and subspace W spanned by w (x-axis)
v = np.array([3, 4, 5])
w = np.array([1, 0, 0])  # subspace = x-axis

# Projection of v onto W
proj_w = (np.dot(v, w) / np.dot(w, w)) * w
# Orthogonal complement (v - projection)
v_perp = v - proj_w

fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

# Plane for W (yz-plane)
y = np.linspace(-6, 6, 10)
z = np.linspace(-6, 6, 10)
Y, Z = np.meshgrid(y, z)
X = np.zeros_like(Y)
ax.plot_surface(X, Y, Z, alpha=0.2, color='cyan')

# Draw vectors
ax.quiver(0, 0, 0, *v, color='r', label='Vector v', linewidth=2)
ax.quiver(0, 0, 0, *proj_w, color='g', label='Projection on W', linewidth=2)
ax.quiver(*proj_w, *v_perp, color='b', label='Orthogonal Component', linewidth=2)

# Labels
ax.set_xlim([-6, 6])
ax.set_ylim([-6, 6])
ax.set_zlim([-6, 6])
ax.set_xlabel('X-axis (W)')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
ax.set_title('Projection and Orthogonal Complement')
ax.legend()

plt.show()
```
