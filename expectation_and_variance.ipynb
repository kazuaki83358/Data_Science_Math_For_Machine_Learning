```python
#Graph of Expectation (Mean)
import numpy as np
import matplotlib.pyplot as plt

X = np.random.normal(5, 2, 5000)  # mean=5, std=2

plt.hist(X, bins=50, density=True, alpha=0.6)
plt.axvline(np.mean(X), color='red', label=f"Mean = {np.mean(X):.2f}")

plt.title("Expectation (Mean)")
plt.legend()
plt.grid()
plt.show()
```

```python
#Graph of Variance (Spread)
import numpy as np
import matplotlib.pyplot as plt

X1 = np.random.normal(0, 1, 5000)   # low variance
X2 = np.random.normal(0, 5, 5000)   # high variance

plt.hist(X1, bins=50, density=True, alpha=0.6, label="Low Variance (σ=1)")
plt.hist(X2, bins=50, density=True, alpha=0.6, label="High Variance (σ=5)")

plt.title("Variance Visualization")
plt.legend()
plt.grid()
plt.show()
```

```python
#Standard Deviation
import numpy as np

X = np.array([1, 2, 3, 4, 5])

expectation = np.mean(X)
variance = np.var(X)
std_dev = np.std(X)

print("E[X] =", expectation)
print("Var(X) =", variance)
print("SD(X) =", std_dev)
```
