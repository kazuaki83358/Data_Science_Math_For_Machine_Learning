```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Sample data (x and y)
x = np.array([1, 2, 3, 4])
y = np.array([2, 2.9, 4.1, 5.05])
```

```python
# Add bias term (for c)
X = np.vstack([x, np.ones(len(x))]).T  # [[1,1], [2,1], [3,1], [4,1]]
```

```python
# ----- Least Squares Solution -----
w_ls = np.linalg.inv(X.T @ X) @ X.T @ y  # (X^T X)^-1 X^T y
m_ls, c_ls = w_ls
```

```python
# ----- Minimum Norm Solution -----
w_min_norm = np.linalg.pinv(X) @ y  # X+ y
m_mn, c_mn = w_min_norm
```

```python
# Predictions
y_pred_ls = m_ls * x + c_ls
y_pred_mn = m_mn * x + c_mn
```

```python
# ----- Plot -----
plt.figure(figsize=(8,6))
plt.scatter(x, y, color='red', label='Data points', s=60)
plt.plot(x, y_pred_ls, label=f'Least Squares Line: y={m_ls:.2f}x+{c_ls:.2f}', color='blue', linewidth=2)
plt.plot(x, y_pred_mn, '--', label=f'Minimum Norm Line: y={m_mn:.2f}x+{c_mn:.2f}', color='green')

plt.title("Least Squares vs Minimum Norm Solution")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()
```
