```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Basis vector for R2
i = np.array([1,0])
j = np.array([0,1])
```

```python
#draw basis vectors 
plt.figure(figsize=(6,6))
plt.quiver(0,0,i[0],i[1],angles='xy',scale_units='xy',scale=1,color='r',label='i = (1,0)')
plt.quiver(0,0,j[0],j[1],angles='xy',scale_units='xy',scale=1,color='b',label='j = (0,1)')

# Set graph limits
plt.xlim(-2, 2)
plt.ylim(-2, 2)
plt.axhline(0, color='black')
plt.axvline(0, color='black')
plt.grid(True)
plt.legend()
plt.title("Basis Vectors of 2D Space (R²)")
plt.show()
```

```python
v1 = np.array([1, 2])
v2 = np.array([2, 4])
```

```python
# Stack as columns
matrix = np.column_stack((v1, v2))
```

```python
rank = np.linalg.matrix_rank(matrix)
print("Rank (Dimension):", rank)
```

```python
#draw basis vectors 
plt.figure(figsize=(6,6))
plt.quiver(0,0,v1[0],v1[1],angles='xy',scale_units='xy',scale=1,color='r',label='v1 = (1,2)')
plt.quiver(0,0,v2[0],v2[1],angles='xy',scale_units='xy',scale=1,color='b',label='v2 = (2,4)')

# Set graph limits
plt.xlim(-2, 2)
plt.ylim(-2, 2)
plt.axhline(0, color='black')
plt.axvline(0, color='black')
plt.grid(True)
plt.legend()
plt.title("Basis Vectors of 2D Space (R²)")
plt.show()
```
