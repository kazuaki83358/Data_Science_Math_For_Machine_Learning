```python
import pandas as pd
df = pd.read_csv("weather_data.csv",parse_dates=["day"])
df.set_index('day',inplace=True)
df
```

```python
new_df = df.fillna(0)
new_df
```

```python
new_df = df.fillna(
    {'temperature' : 0,
     'windspeed' : 0,
     'event' : 'no event'
    }
)
new_df
```

```python
new_df = df.ffill()
new_df
```

```python
new_df = df.ffill(limit=2)
new_df
```

```python
df_numeric = df.select_dtypes(include=['number']).copy()

for col in df_numeric.columns:
    df_numeric[col] = pd.to_numeric(df_numeric[col], errors='coerce')

df_interpolated_numeric = df_numeric.interpolate(method="time")

new_df = df_interpolated_numeric.copy()
new_df['event'] = df.ffill()['event']

new_csv = new_df.to_csv('new_weather_data.csv')
new_df
```

```python
new_df = df.dropna(thresh=1)
new_df
```
