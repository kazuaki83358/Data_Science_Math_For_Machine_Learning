```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define symmetric matrix
A = np.array([[3, 1],
              [1, 3]])
```

```python
# Eigen decomposition
eigvals, eigvecs = np.linalg.eigh(A)
```

```python
print("Eigenvalues:", eigvals)
```

```python
print("Eigenvectors:\n", eigvecs)
```

```python
#Reconstruct A using QÎ›Q^T
Lambda = np.diag(eigvals)
print(Lambda)
A_reconstructed = eigvecs @ Lambda @ eigvecs.T
print("\nReconstructed Matrix:\n", A_reconstructed)
```

```python
# Visualize Transformation
# Original grid
x = np.linspace(-2, 2, 20)
y = np.linspace(-2, 2, 20)
X, Y = np.meshgrid(x, y)
points = np.stack([X.flatten(), Y.flatten()])
```

```python
# Transform points using A
transformed_points = A @ points
```

```python
#Plot
plt.figure(figsize=(8, 8))
plt.quiver(points[0], points[1],
           transformed_points[0] - points[0],
           transformed_points[1] - points[1],
           angles='xy', scale_units='xy', scale=1, color='lightgray', alpha=0.5)
```

```python
# Eigenvectors
origin = np.zeros((2, 1))
for i in range(2):
    plt.quiver(*origin, *eigvecs[:, i]*eigvals[i],
               angles='xy', scale_units='xy', scale=1,
               color=['r', 'b'][i], label=f'Eigenvector {i+1}')

plt.title("Spectral Decomposition Visualization")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()
plt.axis('equal')
plt.grid(True)
plt.show()
```
