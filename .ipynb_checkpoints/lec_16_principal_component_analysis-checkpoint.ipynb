```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
```

```python
# Create synthetic 2D data
np.random.seed(42)
X = np.dot(np.random.rand(2, 2), np.random.randn(2, 200)).T
```

```python
# Step 2: Apply PCA
pca = PCA(n_components=2)
pca.fit(X)
```

```python
# Step 3: Get the principal components
mean = np.mean(X, axis=0)
components = pca.components_
explained_var = pca.explained_variance_
```

```python
# Step 4: Visualize
plt.figure(figsize=(6,6))
plt.scatter(X[:, 0], X[:, 1], alpha=0.3)
for length, vector in zip(np.sqrt(explained_var), components):
    v = vector * 3 * length
    plt.plot([mean[0], mean[0] + v[0]], [mean[1], mean[1] + v[1]], linewidth=3)
plt.axis('equal')
plt.title("PCA — Principal Components (Directions of Max Variance)")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.datasets import load_iris

# 1️⃣ Load data
data = load_iris()
X = data.data   # shape (150, 4)
y = data.target # flower labels (not used in PCA math)

# 2️⃣ Perform PCA
pca = PCA(n_components=2)  # reduce to 2D for easy plotting
X_pca = pca.fit_transform(X)

# 3️⃣ Check how much variance is captured by each component
print("Explained variance ratio:", pca.explained_variance_ratio_)

# 4️⃣ Visualize result
plt.scatter(X_pca[:,0], X_pca[:,1], c=y, cmap='viridis')
plt.xlabel('PC 1')
plt.ylabel('PC 2')
plt.title('PCA on Iris Dataset')
plt.show()
```
