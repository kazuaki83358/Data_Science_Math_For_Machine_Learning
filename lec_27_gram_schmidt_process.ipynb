```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Gram-Schmidt Orthogonalization
# -----------------------------
def gram_schmidt(vectors):
    """Perform Gram-Schmidt process on a list of vectors."""
    orthogonal = []
    for v in vectors:
        u = v.copy()
        for prev_u in orthogonal:
            proj = np.dot(v, prev_u) / np.dot(prev_u, prev_u) * prev_u
            u = u - proj
        orthogonal.append(u)
    return orthogonal
```

```python
# Normalize vectors (for orthonormal basis)
def normalize(vectors):
    return [v / np.linalg.norm(v) for v in vectors]
```

```python
# Example vectors
# -----------------------------
v1 = np.array([2, 1])
v2 = np.array([1, 2])

vectors = [v1, v2]
```

```python
# Apply Gram-Schmidt
orthogonal_vectors = gram_schmidt(vectors)
orthonormal_vectors = normalize(orthogonal_vectors)
```

```python
# Print results
# -----------------------------
print("Original Vectors:")
for v in vectors:
    print(v)

print("\nOrthogonal Vectors:")
for u in orthogonal_vectors:
    print(u)

print("\nOrthonormal Vectors:")
for e in orthonormal_vectors:
    print(e)
```

```python
# Visualization
# -----------------------------
origin = np.zeros(2)

plt.figure(figsize=(6,6))
# Plot original vectors
plt.quiver(*origin, v1[0], v1[1], angles='xy', scale_units='xy', scale=1, color='r', label='v1')
plt.quiver(*origin, v2[0], v2[1], angles='xy', scale_units='xy', scale=1, color='b', label='v2')

# Slight offset for orthogonal vectors (so you can see both)
offset = np.array([0.1, 0.1])

plt.quiver(*(origin + offset), orthogonal_vectors[0][0], orthogonal_vectors[0][1],
           angles='xy', scale_units='xy', scale=1, color='g', label='u1 (orthogonal)')
plt.quiver(*(origin + offset), orthogonal_vectors[1][0], orthogonal_vectors[1][1],
           angles='xy', scale_units='xy', scale=1, color='orange', label='u2 (orthogonal)')
plt.xlim(-1, 3)
plt.ylim(-1, 3)
plt.axhline(0, color='gray')
plt.axvline(0, color='gray')
plt.legend()
plt.title("Gram-Schmidt Process: Original vs Orthogonal Vectors")
plt.grid(True)
plt.show()
```
