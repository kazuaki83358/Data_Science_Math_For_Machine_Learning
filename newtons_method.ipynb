```python
import numpy as np
import matplotlib.pyplot as plt

# Function
def f(x):
    return x**2

# First derivative
def df(x):
    return 2*x

# Second derivative (Hessian)
def d2f(x):
    return 2

# Newton’s Method
x = 5
xs = [x]  # Track points

for i in range(6):
    x = x - df(x) / d2f(x)
    xs.append(x)

# Plotting
x_range = np.linspace(-6, 6, 400)
y_range = f(x_range)

plt.plot(x_range, y_range, label="f(x) = x²")
plt.scatter(xs, [f(i) for i in xs], color="red")

for i, point in enumerate(xs):
    plt.annotate(f"x{i}", (point, f(point)))

plt.title("Newton’s Method Path on f(x)=x²")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid()
plt.legend()
plt.show()
```
