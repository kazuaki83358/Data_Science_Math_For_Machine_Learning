```python
#Binomial Distribution Graph
from scipy.stats import binom
import numpy as np
import matplotlib.pyplot as plt

n, p = 10, 0.4
x = np.arange(0, n+1)
y = binom.pmf(x, n, p)

plt.stem(x, y)
plt.title("Binomial Distribution")
plt.xlabel("k")
plt.ylabel("P(X=k)")
plt.grid()
plt.show()
```

```python
#Poisson Distribution Graph
from scipy.stats import poisson
import numpy as np
import matplotlib.pyplot as plt

lam = 5
x = np.arange(0, 15)
y = poisson.pmf(x, lam)

plt.stem(x, y)
plt.title("Poisson Distribution")
plt.xlabel("k")
plt.ylabel("P(X=k)")
plt.grid()
plt.show()
```

```python
#Geometric Distribution Graph
from scipy.stats import geom
import numpy as np
import matplotlib.pyplot as plt

p = 0.3
x = np.arange(1, 15)
y = geom.pmf(x, p)

plt.stem(x, y)
plt.title("Geometric Distribution")
plt.xlabel("Trials until success")
plt.ylabel("Probability")
plt.grid()
plt.show()
```
