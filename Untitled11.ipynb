```python
from cvxopt import matrix , solvers
```

```python
A = matrix([[-1.0,2.0,-1.0,0.0],[-1.0,3.0,0.0,-1.0]])
#print(A)
#A.size
b = matrix([-2.0,12.0,0.0,0.0])
c = matrix([2.0,3.0])
sol = solvers.lp(c,A,b)
print(sol['x'])
```

```python
A = matrix([[-1.0,2.0,-1.0,0.0],[-1.0,3.0,0.0,-1.0]])
#print(A)
#A.size
b = matrix([-2.0,12.0,0.0,0.0])
c = matrix([2.0,3.0])
sol = solvers.lp(c,A,b,options={'maxiters':2})
print(sol['x'])
```

```python
P = matrix([[2.0,1.0],[1.0,4.0]])
q = matrix([3.0,4.0])
G = matrix([[-1.0,0.0],[0.0,-1.0]])
H = matrix([0.0,0.0])
A = matrix([[2.0],[1.0]])
b = matrix([1.0])
sol = solvers.qp(P,q,G,H,A,b)
print(sol['x'])
```
