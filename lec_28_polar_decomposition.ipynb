```python
import numpy as np
from scipy.linalg import svd, sqrtm
```

```python
# Example matrix
A = np.array([[3, 1],
              [0, 2]])
```

```python
# Step 1: SVD
U_s, Sigma, Vt = svd(A)
```

```python
# Step 2: Compute U and P
U = np.dot(U_s, Vt)  # orthogonal/unitary
P = np.dot(np.dot(Vt.T, np.diag(Sigma)), Vt)  # symmetric positive semi-definite
```

```python
print("Original Matrix A:\n", A)
print("\nOrthogonal/Unitary U:\n", U)
print("\nPositive Semi-Definite P:\n", P)
```

```python
# Verify
print("\nReconstructed A (U*P):\n", np.dot(U, P))
```
